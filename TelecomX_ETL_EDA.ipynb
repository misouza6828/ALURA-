```python
# === EXTRAÇÃO ===
!pip -q install pandas matplotlib seaborn requests

import pandas as pd, numpy as np, requests, io
import matplotlib.pyplot as plt, seaborn as sns

pd.set_option('display.max_columns', None)
plt.rcParams['figure.figsize'] = (8,5)

# URL RAW da API/JSON da TelecomX (Alura)
API_URL = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/main/TelecomX_Data.json"

r = requests.get(API_URL, timeout=60)
r.raise_for_status()
raw = r.json()

def to_dataframe(obj):
    if isinstance(obj, list):
        return pd.DataFrame(obj)
    if isinstance(obj, dict):
        for k, v in obj.items():
            if isinstance(v, list):
                return pd.DataFrame(v)
        return pd.json_normalize(obj)
    raise ValueError("Formato inesperado.")

df_raw = to_dataframe(raw)
print("Shape df_raw:", df_raw.shape)
df_raw.head()
```
